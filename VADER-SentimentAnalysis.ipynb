```python
import pandas as pd
import re
```

```python
main_df = pd.read_csv('./dataset/main-cleaned-trump-tweets.csv', sep=',')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
main_df['score'] = main_df['Tweet Content'].apply(lambda tweet: sid.polarity_scores(tweet))
main_df['compound'] = main_df['score'].apply(lambda score: score['compound'])
main_df['polarity'] = main_df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
main_df.head(10)
```

```python
main_df['polarity'].unique()
```

Print out positive and negative tweets by polarity

```python
i = 0
positives = []
neutrals = []
negatives = []

sortedDF = main_df.sort_values(by=['compound'])
# sortedDF.head()
for i in range(0, sortedDF.shape[0]):
    if sortedDF['polarity'][i] == 'neg':
        negatives.append(sortedDF['Tweet Content'][i])
    elif sortedDF['compound'][i] == 0:
        neutrals.append(sortedDF['Tweet Content'][i])
    else:
        positives.append(sortedDF['Tweet Content'][i])
        
        
        
```

```python
print('Positive Tweets: \n')
i = 0
for tweet in positives:
    i += 1
    print(i, tweet)
    print('\n')
```

```python
print('Negative Tweets: \n')
i = 0
for tweet in negatives:
    i += 1
    print(i, tweet)
    print('\n')
```

```python
print('Neutral Tweets: \n')
i = 0
for tweet in neutrals:
    i += 1
    print(i, tweet)
    print('\n')
```
